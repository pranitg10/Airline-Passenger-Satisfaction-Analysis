```python
def separation(data,margin):
    cat=[]
    num=[]
    for i in data.columns:
        if data[i].nunique()< margin:
            cat.append(i)
        else:
            num.append(i)
    return cat,num
```

```python
separation(df,60)
```

```python
cat,num=separation(df,60)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def outlier_treatment(col):
    q1=df[col].quantile(0.25)
    q3=df[col].quantile(0.75)
    iqr=q3-q1
    lf=q1-1.5*iqr
    uf=q3+1.5*iqr
    outliers=df[(df[col]<lf) | (df[col]>uf)]
    print("Percentage of ouliers in ", col, outliers.shape[0] *100/df.shape[0])
    fig , ax=plt.subplots(1,2, figsize=(6,8))
    ax[0].boxplot(df[col])
    ax[1].hist(df[col])
    plt.suptitle(col)
    plt.show()
    df.loc[(df[col]<lf) | (df[col]>uf), col]=df[col].median()
```

```python
for col in num:
    print(outlier_treatment(col))
    print("++++_________________++++")
```
